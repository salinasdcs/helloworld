```python
import os
import pandas as pd
import numpy as np
import random
```

```python
csvFilePath = os.path.join(os.getcwd(), 'roster.csv')
```

### Read CSV to DataFrame

```python
roster = pd.read_csv(csvFilePath)
print(type(roster))
```

#### Viewing the data

### Modifying the Data

```python
d = {'name': ['Wally']}
tmp_df =pd.DataFrame(data = d)
roster = pd.concat([roster, tmp_df], ignore_index = True)
```

```python
roster.tail()
```

#### Assign Grades

```python
np.random.seed(1)
roster['grade'] = np.random.randint(0, 100, size = len(roster))
roster
```

.loc can be used with a boolean array (i.e. of 1s and 0s)

```python
roster['name'] == "Daniela P"
```

```python
roster.loc[roster['name'] == "Daniela P", 'grade'] = 100
roster
```
